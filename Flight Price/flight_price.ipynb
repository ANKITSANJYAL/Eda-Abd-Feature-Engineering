## Flight Price prediction data(eda and feature engineering)

**IMPORTING REQUIRED LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')    # train data
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')   # test Data
test_df.head()
```

```python
final_df = train_df.append(test_df)   # merging train and test data
final_df.tail()
```

```python
final_df.info()
```

**Here we have a date of journey feature which is in object format so now our task is to convert this thing into integer
so we will split it like date month and year so that we can deal with them individually**

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head(2)
```

```python
final_df['Date'] = final_df['Date'].astype(int)    # converting into integer
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)    # Dropping the date_of_journey column since its of now use now
final_df.head(1)
```

```python
final_df.isnull().sum()   # checking the null valued
```

```python
#Lets deal with arrival time   
#Splitting the arrival time into 2 parts i.e, minutes and hours and saving in integer format
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head(1)
```

```python
final_df['Arrival_hour']= final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_min']= final_df['Arrival_Time'].str.split(':').str[1].astype(int)
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(2)
```

```python
#Lets deal with Depature time   
#Splitting the Depature time into 2 parts i.e, minutes and hours and saving in integer format
```

```python
final_df['Dept_hour'] = final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Dept_min'] = final_df['Dep_Time'].str.split(':').str[1].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
final_df.head(1)
```

```python
# Lets encode the total stops feature
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})

final_df.head()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].fillna(1)
```

```python
final_df['Total_Stops'].unique()
```

**Since the Route column is of no use now coz we already know the number of stops the plane is going to make so we can delete the feature**

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head(1)
```

**Lets change the format of the duration now as the feature is in the format of xh ym where x is number of hours and y is number of minutes**
  
**My idea here is splitting them first and then adding them to give x.y/60 hrs as the final duration**

```python
final_df['duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['duration_hour']=='5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['duration_minutes'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0)
```

```python
final_df['duration_minutes'] = final_df['duration_minutes'].astype(int)
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
```

```python
final_df['duration_minutes'] = final_df['duration_minutes']/60
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df['Duration'] = final_df['duration_hour']+final_df['duration_minutes']
final_df.head()
```

```python
final_df.drop(['duration_hour','duration_minutes'],axis=1,inplace=True)
final_df.head(1)
```

**Now lets encode all the categorical features in the data for that I will be using the sklearn LabelEncoder  And thats not the only option I mean we can surely use the OneHotEncoder or pandas getdummy methods**

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

for columns in final_df.columns:
    if final_df[columns].dtype=='object':
        final_df[columns] = labelencoder.fit_transform(final_df[columns])
final_df.head(2)
        
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

**Since we have made our data ready to use for our machine learning model now lets divide them into train and test parts**  
  
**Here if the data has null value in the Price column that will be taken as the test data and rest of the other will be taken as the train data**

```python
train_df = final_df[~final_df['Price'].isnull()]
test_df = final_df[final_df['Price'].isnull()]
```
